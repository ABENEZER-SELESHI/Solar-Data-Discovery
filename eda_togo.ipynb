```python
# Loop through numerical columns and identify outliers using IQR
numerical_cols = df.select_dtypes(include=['number']).columns

outliers = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers[col] = len(outlier_rows)

# Show number of outliers per column
print("Outlier count per numerical column:")
print(outliers)
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    # Check if the column is truly numeric and has enough data to plot
    if pd.api.types.is_numeric_dtype(df[col]) and df[col].dropna().size > 1:
        plt.figure(figsize=(6, 1))
        sns.boxplot(x=df[col].dropna(), color='skyblue') # Drop NaNs for plotting
        plt.title(f'Boxplot of {col}')
        plt.show()
    else:
        print(f"Skipping boxplot for non-numerical or empty column: {col}")
```
